```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
## Sigmoid
```

```python
def sigmoid(x):
    sigmoid_func =  1/(1+ np.exp(-x))
    return sigmoid_func


def sigmoid_derivative(x):
    sigmoid_derivative_func = sigmoid(x) * (1-sigmoid(x))
    return sigmoid_derivative_func
```

```python
x  = np.linspace(-10,10,50)
y = sigmoid(x)
y_d = sigmoid_derivative(x)

plt.plot(x,y, color='red', label='sigmoid')
plt.plot(x,y_d, color='green', label='sigmoid_derivative')
plt.grid()
plt.legend()
```
