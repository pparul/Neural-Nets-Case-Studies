```python
#Calculate Cosine Similarity Between Vectors

a = [1,2,3]
b = [1,3,4]

cs = 0
mag_a = 0
mag_b = 0

for i in range(len(a)):
    cs = cs + a[i]*b[i]
    mag_a =  mag_a + a[i]**2
    mag_b =  mag_b + b[i]**2

cs = cs / (mag_a**0.5 * mag_b**0.5) 
cs
```
