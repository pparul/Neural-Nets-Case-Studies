```python
def productExceptSelf(nums):
    n = len(nums)
    result = [1] * n
    
    # Left pass: result[i] = product of everything to the left
    prefix = 1
    for i in range(n):
        result[i] = prefix
        prefix *= nums[i]
        #print(prefix)
    
    print(result)
    #print('\n')
    # Right pass: multiply by product of everything to the right
    suffix = 1
    for i in range(n - 1, -1, -1):
        print(i)
        result[i] *= suffix
        suffix *= nums[i]
    
    return result

nums=[2,3,4,6]
result = productExceptSelf(nums)
print(result)
```

```python

def hasDuplicate(nums: list[int]) -> bool:
    duplicate_dict = {}
    returnVal = False
    for i in nums:
            if duplicate_dict.get(i,0):
                returnVal = True
            else:
                 duplicate_dict[i] = 1

    return returnVal

nums = [1,2,3,3]
hasDuplicate(nums)
```

```python
def replaceElements(arr: list[int]) -> list[int]:
    new_arr = [1]*len(arr)
    for i in range(len(arr)-1):
        new_arr[i] = arr[i]
        for j in range(i, len(arr[i:-1])):       
            if arr[i] < arr[j]:
                new_arr[i] = arr[j]
                break
    new_arr[-1] = -1
    return new_arr  

arr=[2,4,5,3,1,2] 
replaceElements(arr)
```

```python
def subarraySum(nums: list[int], k: int) -> int:
    count = 0
    prefix_sum = 0
    prefix_map = {0: -1}  # sum -> how many times we've seen it
    
    for idx, num in enumerate(nums):
        prefix_sum += num
        if (prefix_sum - k) in prefix_map:
            count = count+1
        print(f'{prefix_sum}, {count}')
        prefix_map[prefix_sum] = idx
        print(prefix_map)
        print('\n')
    
    return count
nums = [1, -1, 1, -1, 1]
k = 0
subarraySum(nums, k)
```
